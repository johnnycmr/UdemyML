```python
salarios_marcos=np.array([100,200,400,300])
salarios_gilenos = np.array([50,300,500,450])
```

```python
plt.plot(salarios_marcos,c='Black',ls='--',marker='s',label='Salarios Marcos')
plt.plot(salarios_gilenos,c='Red',ls='--',marker='s',label='Salarios Gileno')
plt.show()
```

```python
plt.plot(salarios_marcos,c='Black',ls='--',marker='s',label='Salarios Marcos')
plt.plot(salarios_gilenos,c='Red',ls='--',marker='^',label='Salarios Gileno')
plt.legend()
plt.show()
```

```python
salarios_pedro=np.array([200,150,300,700])
plt.plot(salarios_marcos,c='Black',ls='--',marker='s',label='Salarios Marcos')
plt.plot(salarios_gilenos,c='Red',ls='--',marker='s',label='Salarios Gileno')
plt.plot(salarios_pedro,c='Green',ls='--',marker='o',label='Salarios Pedro')
plt.legend()
plt.show()
```
